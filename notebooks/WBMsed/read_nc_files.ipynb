```python
datasets = Path("/work/pi_kandread_umass_edu/Datasets/")
save_dir = Path("/work/pi_kandread_umass_edu/tss-ml/data/NA_WQP/")
preprocess_dir = save_dir / "preprocess"

matchups = pd.read_csv(
    preprocess_dir / "hydrography_matchups.csv",
    index_col='wqp_id',
    converters={"wqp_multi_ids": literal_eval})
matchups['geometry'] = matchups['geometry'].apply(wkt.loads)

# Create a GeoDataFrame
matchups = gpd.GeoDataFrame(matchups, geometry='geometry')
matchups.set_crs("ESRI:102004", inplace=True)

matchups_wgs84 = matchups.to_crs(epsg=4326)
```

```python
# Extract latitude and longitude from the geometry column
matchups_wgs84['lat'] = matchups_wgs84.geometry.y
matchups_wgs84['lon'] = matchups_wgs84.geometry.x

# Select all relevant points at once
selected_points = ds['SedimentFlux'].sel(
    latitude=xr.DataArray(matchups_wgs84['lat'], dims='points'),
    longitude=xr.DataArray(matchups_wgs84['lon'], dims='points'),
    method='nearest'
).drop_vars(['latitude', 'longitude'])
```
