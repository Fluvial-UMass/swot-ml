# sbatch dir in swot-ml

This notebook should very briefly show you basics of running the model training from command line using the bash scripts. These are in the `sbatch` directory of the swot-ml repo and are written with some assumptions about the HPC and accounts. You probably should review the job arguments they use. Their purpose is to make it really easy to run sbatch job(s) based on the config files and a few arguments. There are a few scripts, but the ones I use regularly now are `run.sh` and `run_dir.sh`.


Normally I do this from a command line, but easier to document it in a notebook here.

```python
!/nas/cee-water/cjgleason/ted/swot-ml/sbatch/run.sh
```

```python
!/nas/cee-water/cjgleason/ted/swot-ml/sbatch/run_dir.sh
```

As their names imply, `run.sh` trains a single model based on a config file and `run_dir.sh` will call `run.sh` on all yml files found within a directory. I won't actually run them here since it takes a lot of resources, but once you have a yml file or directory of files you would call them like this:

```python
# This gives us a weird error, but it has checked for a file matching that path and didn't find anything
# If you replace this with an actual .yml file path it should run. 
!/nas/cee-water/cjgleason/ted/swot-ml/sbatch/run.sh gpupod train your_training_config.yml
```

```python
# run_dir does a similar check looking for .yml files, and then will prompt you to confirm with a 'y' or 'n' option. 
# Each one of these will spin up a new gpupod job, so it's good to check we aren't accidentally sending a truckload of training configs accidentally. 
!/nas/cee-water/cjgleason/ted/swot-ml/sbatch/run_dir.sh gpupod train /nas/cee-water/cjgleason/ted/swot-ml/runs/reservoir_sf_grnn

# I'm not actually sure how to respond with 'n', so you may just have to kill the cell here to break out of the waiting for a response... 
```
