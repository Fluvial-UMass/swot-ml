```python
import xarray as xr
import pandas as pd
import numpy as np
import os
import shutil
from tqdm import tqdm

from data import ZarrDataStore
```

```python
save_file = "zarr_test.zarr"

basins = [str(i) for i in range(100)]
t1 = "1980-01-01"
t2 = "2024-12-31"
date_range = pd.date_range(start=t1, end=t2, freq='D')

zds = ZarrDataStore(save_file, basins, date_range, overwrite=True)
```

```python
data = np.random.randint(0, 10, (len(basins), len(date_range), 2))
data.shape
```

```python
for i, b in tqdm(enumerate(basins)):
    src1 = pd.DataFrame(index=date_range, columns=['test1','test2'], data=data[i,...])
    zds.add_dynamic_source(b, 'src2', src1)
```
